```python
tickers=["AAPL","AMZN","MSFT","NVDA","TSLA","SPY","NVDA","FSLR","MSFT"]
```

```python
ohlc_intra={} #directory with ohlc value for each stock
start=dt.datetime.today()-dt.timedelta(days=1)
end=dt.datetime.today()

for ticker in tickers:
    ohlc_intra[ticker]=yf.download(ticker,period='1d',interval='5m')
    ohlc_intra[ticker].dropna(inplace=True,how="all")
    ohlc_intra[ticker]=ohlc_intra[ticker].iloc[:-1,:]

tickers=ohlc_intra.keys()
```

```python
df_dict={}
for ticker in tickers:
    df_dict[ticker]=yf.download(ticker,period='2d',interval='5m')
    df_dict[ticker]=df_dict[ticker].iloc[:-1,:]
    df_dict[ticker]["ema_9"]=ta.trend.EMAIndicator(df_dict[ticker]["Close"],9,False).ema_indicator()
    df_dict[ticker]["ema_21"]=ta.trend.EMAIndicator(df_dict[ticker]["Close"],21,False).ema_indicator()
    df_dict[ticker]=pd.merge(left=df_dict[ticker],right=ohlc_intra[ticker],how='inner')
    
```

```python
for ticker in tickers:
    df_dict[ticker].index=ohlc_intra[ticker].index
```

```python
for ticker in tickers:
    trend=[]
    for i in range(len(ohlc_intra[ticker])):
        #condition of uptrend
        if (ohlc_intra[ticker]['Open'][i]>ohlc_intra[ticker]['Open'][i-1] and ohlc_intra[ticker]['Close'][i]>ohlc_intra[ticker]['Close'][i-1]) and \
        (ohlc_intra[ticker]['Open'][i-1]>ohlc_intra[ticker]['Open'][i-2] and ohlc_intra[ticker]['Close'][i-1]>ohlc_intra[ticker]['Close'][i-2]):
            if ohlc_intra[ticker]['Open'][i]>df_dict[ticker]['ema_9'][i]:
                trend.append(1)
            else:
                trend.append(0)


        #condition of downtrend
        elif (ohlc_intra[ticker]['Open'][i]<ohlc_intra[ticker]['Open'][i-1] and ohlc_intra[ticker]['Close'][i]<ohlc_intra[ticker]['Close'][i-1]) and \
        (ohlc_intra[ticker]['Open'][i-1]<ohlc_intra[ticker]['Open'][i-2] and ohlc_intra[ticker]['Close'][i-1]<ohlc_intra[ticker]['Close'][i-2]):
            if ohlc_intra[ticker]['Open'][i]<df_dict[ticker]['ema_9'][i]:
                trend.append(-1)
            else:
                trend.append(0)
        else:
            trend.append(0)
        
    ohlc_intra[ticker]['Trend']=trend
    ohlc_intra[ticker]['Cum_Trend']=ohlc_intra[ticker]['Trend'].cumsum()
        
    
```

```python
#Dataframe method
df=pd.DataFrame()
for ticker in tickers:
    df[ticker]=ohlc_intra[ticker]['Cum_Trend']
```
