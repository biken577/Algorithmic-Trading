```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import mplfinance as mpf
import datetime as dt
```

```python
dt.datetime(2022,10,28)
```

```python
end=dt.datetime(2022,10,28)-dt.timedelta(days=0)
end
```

```python
start=end-dt.timedelta(days=1)
start
```

```python
spy=yf.download('spy',start=start,end=end,interval='5m')
```

```python
mpf.plot(spy,type='candle')
```

```python
mpf.plot(spy,type='candle',mav=(9,21,50),volume=True)
```
